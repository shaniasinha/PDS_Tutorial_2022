```
# Check for null values in the dataset
df.isnull().sum()
```

```
# Check what fraction of all observations is null
for i in df.columns:
    null_rate = df[i].isna().sum() / len(df) * 100
    if null_rate > 0.0:
      print("{}: {}%".format(i,round(null_rate,2)))
```

From the results above, we can see that the variable **director** has the highest number of null values. 

**cast** and **country** have similar number of missing (or null) values.

## Cleaning the dataset

Here, we get to make the choice of what we want to keep and what we want to remove from our dataset. 

I am making the following choices:

*Remove all rows where **date_added** is null.* 
> **date_added** refers to the date when the media content was added to Netflix. This information is useful for our analysis but the number of missing observations compared to the size of the population is very small.


*Remove all rows where **rating** is null.* 
> **rating** refers to the how well-liked the media content is by the audience. This information is useful for our analysis but the number of missing observations compared to the size of the population is very small.


*Remove all rows where **duration** is null.*
